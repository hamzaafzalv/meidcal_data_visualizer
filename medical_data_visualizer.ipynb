```python
import pandas as pd
```

```python
df =pd.read_csv('medical_examination.csv')
```

```python
df
```

```python
overweight = ''
df['overweight'] = overweight
df
```

```python
height_meter = df['height']/100
height_meter
df['height'] = height_meter
df
```

```python
BMI = df['weight']/df['height']*df['height'] #BMI
BMI
df['BMI'] = BMI
df
```

```python
df.loc[df['BMI'] > 25, 'overweight'] = 1
df.loc[df['BMI'] < 25, 'overweight'] = 0
df
```

```python
df
```

```python
df.loc[(df['gluc'] == 1) ,'gluc'] = 0
df
```

```python
df.loc[(df['gluc'] > 1) ,'gluc'] = 1
df
```

```python
df.loc[(df['cholesterol'] == 1) ,'cholesterol'] = 0
df.loc[(df['cholesterol'] > 1) ,'cholesterol'] = 1
df

```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_cat = pd.melt(df, id_vars = ['cardio'], value_vars = ['cholesterol','gluc','smoke','active','alco','overweight'])
df_cat
```

```python
df_cat['total'] = 1
df_cat=df_cat.groupby(['cardio','variable','value'],as_index=False).count()
```

```python
import seaborn as sns
```

```python
fig = sns.catplot(x='variable',y='total',data = df_cat,hue='value',kind = 'bar',col='cardio').fig
```

```python
fig.savefig('catplot.png')
fig

```

```python
df_heat= df[   
(df['ap_lo'] <= df['ap_hi']) & 
(df['height'] >= df['height'].quantile(0.025))&
(df['height'] <= df['height'].quantile(0.975)) &
(df['weight'] >= df['weight'].quantile(0.025)) &
(df['weight'] <= df['weight'].quantile(0.975))
]
```

```python
corr = df_heat.corr(method='pearson')
mask = np.triu(corr)
fig,ax =  plt.subplots(figsize=(12,12))
sns.heatmap(corr,linewidth=1,annot=True,square=True,mask=mask,fmt='.1f',center= 0.08,cbar_kws={'shrink':0.5})
```
